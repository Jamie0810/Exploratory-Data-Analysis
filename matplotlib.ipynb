```python
import matplotlib.pyplot as plt

steak_price = [14.3, 15, 17.1, 19.2, 21]
pork_chops = [9.3, 10.1, 10.9, 11.2, 12]
year = [2001, 2002, 2003, 2004, 2005]


# Create a line chart
plt.plot(year, steak_price)
plt.plot(year, pork_chops, 'v')
# plt.plot(year, steak_price, year, pork_chops)

# https://matplotlib.org/3.5.1/api/_as_gen/matplotlib.pyplot.plot.html#:~:text=may%20be%20ambiguous.-,Markers,-character

# plt.scatter(year, pork_chops)

# Show the chart
plt.show()
```

```python
# Add some attributes
plt.plot(year, steak_price, label='Steak')
plt.plot(year, pork_chops, label='Pork')
plt.legend()

# Styling
# plt.style.use('seaborn')
# https://problemsolvingwithpython.com/06-Plotting-with-Matplotlib/06.13-Plot-Styles/


# x,y labels
plt.xlabel('Year')
plt.ylabel('Steak Price/1 kilogram')

# Customize axis
plt.axis([2000, 2006, 5, 25])

# plot title
plt.title("Steak & Pork Yearly Price")

plt.savefig('Steak & Pork Yearly Price.jpg')
plt.show()
```
